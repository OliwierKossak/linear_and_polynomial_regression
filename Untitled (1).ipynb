```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

data_df = pd.read_csv('cube_3x3.csv')

data_df.rename(columns={'No.': 'solution_number', 'Time': 'solution_time_seconds'}, inplace=True)

data_df['solution_time_seconds'] = pd.to_numeric(data_df['solution_time_seconds'], errors='coerce')
data_df['solution_number'] = data_df['solution_number'].astype(int)

data_df.dropna(inplace=True, axis=0)

X = np.array(data_df['solution_number']).reshape(-1, 1)
y = np.array(data_df['solution_time_seconds']).reshape(-1, 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

sc = StandardScaler()
sc.fit(X_train)

X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

lr = LinearRegression()

lr.fit(X_train_std, y_train)

y_pred = lr.predict(X_test_std)

plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred, color='magenta', linewidth=1)
plt.xlabel('solution number')
plt.ylabel('solution time (sec)')
plt.legend(['data','linear regression'])
plt.show()

```

```python
poly_second_degree = PolynomialFeatures(degree=2, include_bias=False)

# Fit to data, then transform it.
x_new = poly_second_degree.fit_transform(X_train)
x_new_test = poly_second_degree.fit_transform(X_test)

model_lr = LinearRegression()

model_lr.fit(x_new, y_train)

poly_model = np.poly1d(np.polyfit(np.array(data_df['solution_number']), np.array(data_df['solution_time_seconds']), 2))
poly_line = np.linspace(1, 3000, 2661)
plt.scatter(X_test, y_test)
plt.plot(poly_line, poly_model(poly_line), color='red')
plt.xlabel('solution number')
plt.ylabel('solution time (sec)')
plt.legend(['data', 'polynomial regression=2'])
plt.show()

linear_model_score = round(lr.score(X_test_std, y_test), 2)
poly_model_score = round(model_lr.score(x_new_test, y_test), 2)

print('Linear Regression score:', linear_model_score)
print('Polynomial Regression 2 degreee score:', poly_model_score)
```
